## Step 1: Open the exe file

### This is the bolded row in the folder. Try to click on the icon
![open](open.png)

## Step 2: Input the Category of news you want to download 

### Please type a number inside with where each number corresponds to the categories documented below
![input1](input1.png)

## Step 3: Input the pages of news you want to download

### Please insert the number in forms of a-b (e.g. 2-3) which means the program will extract the TIMES news from page 2 to page 3.

### If you just want to extract one page, please type a-a(e.g. 1-1)

### The page is defined according to the official website of TIMES

![page](page.png)

### Insert C:\Users\... as the directory location
