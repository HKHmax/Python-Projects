## Step 1: Get the HTML code from the website (start with a particular sticker page: https://whatsticker.online/p/319193X6T54NB/HK/zh) 
### The web is in the format of the following:
![Whatapp Sticker](3.png)

```python
import requests
url = 'https://whatsticker.online/p/319193X6T54NB/HK/zh'
html = requests.get(url)
#check if the request fromt the website is successful
if html.status_code == requests.codes.ok:
    print(html.text)
```

```python
#save the html code into the variable html_code
html_code = html.text
#see if we successfully save it
html_code
```

## Step 2: Use BeautifulSoup to analysis the HTML code structure and get what we what
### This usually involves two big steps:
1. Analyse the HTML of the web in great details
2. Use code to extract the useful parts

```python
from bs4 import BeautifulSoup
#first encode the html_code to utf-8
html.encoding = 'UTF-8'
#put the html into the BS and analyse it with lxml method
sp = BeautifulSoup(html_code, 'lxml')
```

```python
#look at the title of the sp
print(sp.title)
```

```python
#header of web
print(sp.h1)
```

```python
#paragraph of sp
print(sp.p)
```

### After analysing the HTML code, I find that the whatsapp stickers are under the class called "sticker"(below the class 'sticker-list") and its link is at the  <source>  above it as shown in the below graph and :
![Whatapp Sticker](5.png)

```python
#get the stickers position 
sticker = sp.find_all('source', {'type':'image/webp'})
sticker
```

```python
#get the src inside the class
for data in sticker:
    src = data.get('srcset')
    print(src)
```

```python
#create directory for saving the pictures
images_dir = 'whatsapp_sticker'
#create it if it does not exist
import os
if not os.path.exists(images_dir):
    os.mkdir(images_dir)
```

```python
#then downlaod the stickers with for loop
n = 1
for data in sticker:
    src = data.get('srcset')
    imgfile = requests.get('https://whatsticker.online/' + src)
    #create the full path for saving
    full_path = os.path.join(images_dir, str(n))
    #save the images to the full_path
    with open(full_path + '.png', 'wb') as f:
        f.write(imgfile.content)
    n = n + 1
    print(full_path + '.png')
```

## Step 3: I would like to make things more general to download all the stickers in the main pages

```python
url = 'https://whatsticker.online/c/home/HK/zh'
main_page = requests.get(url)
'pack-item'
#check if the request fromt the website is successful
if main_page.status_code == requests.codes.ok:
    print(main_page.text)
```

```python
#define variable for the html code of main page
main_html = main_page.text
main_html
```

```python
#encode the HTML with UTF-8
main_page.encoding = 'UTF-8'
#put the html into the BS and analyse it with lxml method
sp = BeautifulSoup(main_html, 'lxml')
```

### After studying the HTML code inside, the link of each sticker page is saved at the hyperlink < a > with class = 'pack-item'
![Main Page](7.png)

```python
#get the items with class = 'pack-item'
#get the stickers position 
sticker = sp.find_all('a', {'class':'pack-item'})
sticker
```

```python
#get the link of the respective sticker page
links = []
for data in sticker:
    link = data.get('href')
    links.append(link)
links
```

```python
#go to specific sticker page to crawl all the stickers
m = 1
for link in links:
    html = requests.get(link)
    html.encoding = 'UTF-8'
    html_code = html.text
    sp = BeautifulSoup(html_code, 'lxml')
    sticker = sp.find_all('source', {'type':'image/webp'})
    images_dir = 'whatsapp_sticker_' + str(m)
    if not os.path.exists(images_dir):
        os.mkdir(images_dir)
    #then downlaod the stickers with for loop
    n = 1
    for data in sticker:
        src = data.get('srcset')
        imgfile = requests.get('https://whatsticker.online/' + src)
        #create the full path for saving
        full_path = os.path.join(images_dir, str(n))
        #save the images to the full_path
        with open(full_path + '.png', 'wb') as f:
            f.write(imgfile.content)
        n = n + 1
        print(full_path + '.png')
    m = m + 1
```
